```python
missing_rows = df[df['Open'].isnull()]
missing_rows
```

```python
# Remove such missing values
df_cleaned= df.dropna(subset=['Open'])
df_cleaned.head(30)
```

**Indexes are also removed along with the removed values, if reset_index function not given (Notice index 26)**

```python
df_cleaned= df.dropna(subset=['Open']).reset_index(drop=True)
df_cleaned.head(30)
```

**While ranking our stock prices Null Values(days where there was no trade of that stock, or any other reason) of open high low close are not as useful so we will drop them**

```python
duplicates=df_cleaned.duplicated()
true_count = duplicates.sum()
print(true_count)
# No duplicates
```
